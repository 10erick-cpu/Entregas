# **II.** **Métodos para guardar la información de una componente imagen a un archivo**

##**Redimensionamos una imagen y lo guardamos**




```python
imagen1 = Image.open("joaquin.jpeg")

width, height = imagen1.size

print(imagen1.size)

img = imagen1.resize((450, 450))  # resize the imagen1

plt.imshow(img)
plt.show()
img.save("resize_joaquin.png")  # Save the image (450, 450)
```

# **Métodos para graficar o modificar una componente imagen**

```python
# Obtaining the RGB values on the image for a certain value (x, y)

imagen = Image.open('joaquin.jpeg')

#R,G,B = imagen.getpixel((5,5))
#print(R,G,B)

# Image conversion to numpy array for convenience

image_array = np.array(imagen) 

# print(image_array)


# RGB Channels 

# Channel R

image_array[ : , : , 0]

# Channel G

image_array[ : , : , 1]

# Channel B

image_array[ : , : , 2]
       
# Save Modified Image - Channel R

Image.fromarray(image_array[ : , : , 0]).save('Modified Image 0.png')

# Graphics on the  R - G channels

plt.figure(figsize=(20,20))
plt.subplot(131)
plt.imshow(imagen)
plt.title('Original Image'),plt.axis('on') 

plt.subplot(132)
plt.imshow(image_array[ : , : , 0])
plt.title('Modified Image 0'),plt.axis('off')


plt.subplot(133)
plt.imshow(image_array[ : , : , 1])
plt.title('Modified Image 1'),plt.axis('off')

plt.show()
```
